```python
import nltk
```

```python
#defining features
def word_features(word):
    feature = dict(starting_letter = word[0],last_letter = word[-1:],last2_letters = word[-2:],last3_letters = word[-3:])
    return feature
```

```python
dictionary = open('English_dictionary.txt','r',encoding='utf8').readlines()
```

```python
dictionary
```

```python
#labeling words
labeled_words = list()
pos=0 # assuming the word is located only at 0
label=''
for sentence in dictionary:
    words = nltk.word_tokenize(sentence)
    if len(words)>2:
        pos = 0
        got_label = False
        while not got_label and pos<3:
            if words[pos+1].lower() in '-n.':
                label = 'noun'
                got_label = True
            elif words[pos+1].lower() in '-v.':
                label = 'verb'
                got_label = True
            elif words[pos+1].lower() in '-adj.':
                label = 'adjective'
                got_label = True
            elif words[pos+1].lower() in '-adv.':
                label = 'adverb'
                got_label = True
            else:
                got_label = False
                pos+=1
        labeled_words.append((words[0],label))
        print('*',end='')
```

```python
labeled_words
```

```python
import random
```

```python
random.shuffle(labeled_words)
```

```python
labeled_words
```

```python
featuresets = [(word_features(n), word) for (n, word) in labeled_words]
```

```python
featuresets
```

```python
set_size = len(featuresets)
```

```python
train_set, test_set = featuresets[set_size//2:], featuresets[:set_size//2]
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
print(nltk.classify.accuracy(classifier, test_set))
```
